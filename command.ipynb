```python
# handwrittingtest
# 制作训练集，读trainingDigits文件
import os
from knn import kNN
import numpy as np

def files2matrix(filesPath):
    # path = filesPath #".\\MLiA_SourceCode\\machinelearninginaction\\Ch02\\digits\\trainingDigits"
    files_name = os.listdir(filesPath) # 获取path文件夹中的所有子文件名列表
    # files_name
    files_number = len(files_name)
    feature_matrix = np.zeros((files_number,32*32))
    labels = np.zeros(files_number)
    for i in range(files_number):
        filepath = filesPath+"\\"+files_name[i]
        feature_matrix[i,:] = kNN.image2vector(filepath)
        files_name[i] = files_name[i].replace(".txt","")
        label_number,filecount = files_name[i].split('_')
        labels[i] = label_number
    return feature_matrix,labels

train_path = ".\\MLiA_SourceCode\\machinelearninginaction\\Ch02\\digits\\trainingDigits"
train_feature_matrix , train_labels = files2matrix(train_path)
test_path = ".\\MLiA_SourceCode\\machinelearninginaction\\Ch02\\digits\\testDigits"
test_feature_matrix , test_labels = files2matrix(test_path)
classify_right = kNN.kNNTest(
    train_data=train_feature_matrix, 
    test_data=test_feature_matrix, 
    train_labels=train_labels, 
    test_labels=test_labels,
    k=3)
print("the wrong rate is :",classify_right)
```
